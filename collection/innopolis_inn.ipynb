Импортируем необходимые библиотеки

```python
import requests 
from bs4 import BeautifulSoup
import time
import pandas as pd
```

Получаем содержимое страницы

```python
resp = requests.get('https://sezinnopolis.ru/residents/?arrFilter_pf%5BSTATUS%5D%5B%5D=3&set_filter=Y')
```

Проверяем статус ответа

```python
resp
```

Считываем содержимое html-кода страницы

```python
main_soup = BeautifulSoup(resp.text, 'html.parser')
```

Извлекаем все объекты с классом item

```python
items = main_soup.find_all(class_='item')
```

Извлекаем все ссылки на страницы компаний-резидентов Иннополиса

```python
urls = ['https://sezinnopolis.ru' + item.find('a').get('href') for item in items]
```

Посмотрим на собранный список ссылок

```python
urls
```

С каждой страницы собираем название компании и сайт 

```python
innopolis_companies = list()
innopolis_sites = list()

for url in urls:
    one_site = requests.get(url)

    print(one_site.status_code)

    if one_site.status_code == 200:
        site_soup = BeautifulSoup(one_site.text, 'html.parser')

        comp_name = site_soup.find('h1').text
        comp_site = [a.get('href') for a in site_soup.find(class_='props').find_all('a')][-1]
        innopolis_companies.append(comp_name)
        innopolis_sites.append(comp_site)
        print('Done')

    time.sleep(3)
```

Вручную заполняем список ИНН компаний, данные взяты с ФНС

```python
innopolis_inns = [7703807270, 1615012605, 1660146230, 1624014670, 1624016042, 
                  1615013662, 1615013359, 7728742674, 1655093079, 1615012838, 
                  1615013020, 1615013334, 1615013165, 1615012203, 1615012997, 
                  1615013380, 1655376609, 1650365272, 1615013655, 1615013126
                  ]
```

Создаём список списков, из которого создадим датафрейм

```python
innopolis_tab = [innopolis_companies, innopolis_sites, innopolis_inns]
```

Создаём датафрейм из собранных данных и переименовываем колонки

```python
innopolis = pd.DataFrame(innopolis_tab).T
innopolis = innopolis.rename(columns={0: 'name', 1: 'site', 2: 'inn'})
innopolis
```

Сохраняем результаты в csv

```python
innopolis.to_csv('innopolis_companies.csv')
```
